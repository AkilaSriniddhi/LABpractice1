```python
import matplotlib.pyplot as plt

def plot_case(case_num, n, best, avg, worst):
    labels = ['Best O(1)', 'Average O(n/2)', 'Worst O(n)']
    comparisons = [best, avg, worst]
    colors = ['green', 'orange', 'yellow']

    plt.bar(labels, comparisons, color=colors)
    plt.title(f"Test Case {case_num}: Comparisons for n = {n}")
    plt.ylabel("Comparisons")
    plt.tight_layout()
    plt.show()

case_num = 1

while True:
    s = input("Enter input size (or 'done' to stop): ")
    if s.lower() == 'done':
        print("Exiting...")
        break
    if s.isdigit() and int(s) > 0:
        n = int(s)
        best_c = 1
        avg_c = n // 2 + 1
        worst_c = n

        plot_case(case_num, n, best_c, avg_c, worst_c)
        case_num += 1
    else:
        print("❗ Please enter a valid positive number.")
```
