```python
import time, random, matplotlib.pyplot as plt

def binary_search(arr, x, low, high):
    if low > high: return -1
    mid = (low + high) // 2
    if arr[mid] == x: return mid
    return binary_search(arr, x, low, mid-1) if x < arr[mid] else binary_search(arr, x, mid+1, high)

def time_search(n):
    arr = list(range(n))
    x = random.choice(arr + [n+1])
    t1 = time.perf_counter()
    binary_search(arr, x, 0, n-1)
    return time.perf_counter() - t1

sizes = [ 500, 1000, 5000, 10000, 20000, 50000, 100000, 200000]
times = [time_search(n) for n in sizes]

plt.bar(sizes, times,color='purple',width=3000)
plt.title("Recursive Binary Search: Time vs n")
plt.xlabel("Number of Elements (n)")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.show()
```
