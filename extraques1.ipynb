```python
import random, time, matplotlib.pyplot as plt

def linear_search(arr, key):
    for i, v in enumerate(arr):
        if v == key:
            return i
    return -1

def measure(n):
    arr = random.sample(range(n*10), n)
    times = []
    for key in [arr[0], arr[n//2], -1]:
        start = time.time()
        linear_search(arr, key)
        times.append((time.time() - start) * 1000)
    return times

def plot(times, n):
    colors = ['orange', 'yellow', 'purple']
    plt.pie(times, labels=['Best O(1)', 'Avg O(n/2)', 'Worst O(n)'], autopct='%1.2f%%', colors=colors)
    plt.title(f'Linear Search Times for n={n} (ms)')
    plt.show()

while True:
    n = input("Enter list size n (or 'exit'): ")
    if n.lower() == 'exit': break
    if not n.isdigit() or int(n) <= 0: continue
    n = int(n)
    t = measure(n)
    print(f"Times (ms): Best={t[0]:.5f}, Avg={t[1]:.5f}, Worst={t[2]:.5f}")
    plot(t, n)
```

```python
import matplotlib.pyplot as plt

sizes = []
best = []
avg = []
worst = []

while True:
    s = input("Enter input size (or 'done'): ")
    if s.lower() == 'done': break
    if s.isdigit() and int(s) > 0:
        n = int(s)
        sizes.append(n)
        best.append(1)          # Best case: first element
        avg.append(n // 2 + 1)  # Average case
        worst.append(n)         # Worst case: not found

# Plot
x = range(len(sizes))
w = 0.25

plt.bar([i - w for i in x], best, width=w, color='purple', label='Best O(1)')
plt.bar(x, avg, width=w, color='orange', label='Average O(n/2)')
plt.bar([i + w for i in x], worst, width=w, color='gold', label='Worst O(n)')

plt.xticks(x, sizes)
plt.xlabel("Input Size (n)")
plt.ylabel("Comparisons")
plt.title("Linear Search Comparisons")
plt.legend()
plt.tight_layout()
plt.show()
```
